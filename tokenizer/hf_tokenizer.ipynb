```python
from datasets import load_dataset

dataset_en = load_dataset("wikitext", name="wikitext-2-raw-v1", split="train")
dataset_hi = load_dataset("ai4bharat/sangraha", data_dir="verified/hin", split="train")
dataset_ka = load_dataset("ai4bharat/sangraha", data_dir="verified/kan", split="train")

print(dataset_en[0])
print(dataset_hi[0])
print(dataset_ka[0])
```

```python
dataset_en = [row for row in dataset_en if row['text']]

print(len(dataset_en))
print(len(dataset_hi))
print(len(dataset_ka))
```

```python
import random
import os

# Set the random seed for reproducibility
random.seed(42)
num_samples = 5000  # per language

if not os.path.exists("hf_tokenizer_corpus.txt"):
    with open("hf_tokenizer_corpus.txt", "w", encoding="utf-8") as f:
        for dataset in [dataset_en, dataset_hi, dataset_ka]:
            # Get the length of the dataset
            dataset_length = len(dataset)
            print(f"Dataset length: {dataset_length}")

            # Randomly sample indices
            sampled_indices = random.sample(range(dataset_length), num_samples)
            print(f"Sampled indices: {sampled_indices[:10]}...")  # Print first 10 indices

            # Write sampled texts to file
            for index in sampled_indices:
                f.write(dataset[index]["text"] + "\n")
```
